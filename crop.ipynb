```python
# read all the csv file in the folder
import glob
import numpy as np
left_cors = []
right_cors = []
upper_cors = []
lower_cors = []
for f_name in glob.glob("./data/p1/" + '*.csv'):
    bb_data = pd.read_csv(f_name)
    left_cors.append(bb_data['XM'][0].item() - bb_data['Width'][0].item()/2)
    right_cors.append(bb_data['XM'][0].item() + bb_data['Width'][0].item()/2)
    upper_cors.append(bb_data['YM'][0].item() - bb_data['Height'][0].item()/2)
    lower_cors.append(bb_data['YM'][0].item() + bb_data['Height'][0].item()/2)
```

```python
print(bb_data['XM'][0])
```

```python
minimal_cors = {}
minimal_cors['left'] = min(left_cors)
minimal_cors['right'] = max(right_cors)
minimal_cors['upper'] = min(upper_cors)
minimal_cors['lower'] = max(lower_cors)

print(minimal_cors)
```

```python
fig, ax = plt.subplots(1)
ax.imshow(img)
pat = pa.Rectangle([minimal_cors['lower'], minimal_cors['left']],
                      minimal_cors['right'] - minimal_cors['left'],
                     minimal_cors['lower'] - minimal_cors['upper'])
ax.add_patch(pat)
plt.show()
```
